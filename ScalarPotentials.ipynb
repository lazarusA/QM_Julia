```julia
function step_transmission(Energy, V0)
    if Energy < V0
        return 0.0
    else 
        k1 = sqrt(Energy)
        k2 = sqrt(Energy - V0)
        return k2*abs(2*k1/(k1 + k2))^2 / k1
    end
end
```

```julia
using Plots
energias = linspace(0, 1, 100);
```

```julia
plotly()
```

```julia
plot(energias, step_transmission.(energias, .1), color = :"#0099cc", lw = 2, label = "T",
    size = (400, 300), xlabel = "E", title = "Step Potential")
plot!(energias, 1- step_transmission.(energias, .1), color = :red, lw = 2, label = "R")
```
